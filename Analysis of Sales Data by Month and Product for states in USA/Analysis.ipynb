```python
df["Order Date"]=pd.to_datetime(df["Order Date"],errors="coerce")
df['Month'] = pd.DatetimeIndex(df['Order Date']).month
df[df.isna().any(axis=1)]
```

Seems like entire rows are Nan hence Dropping them

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Quantity Ordered"]=pd.to_numeric(df["Quantity Ordered"])
df["Price Each"]=pd.to_numeric(df["Price Each"])
```

```python
df=df.sort_values("Order Date")
df.head()
```

```python
def get_city(s):
    return s.split(",")[1]+" ("+s.split(",")[2][1:3]+")"
```

```python
df["Sales"]=df["Price Each"]*df["Quantity Ordered"]
df["City"]=df["Purchase Address"].apply(lambda x: f"{get_city(x)}")
df.head()
```

```python
tdf=df.groupby(['Product',"Month"],as_index=False).sum()
```

# Sales by Month for each Product

```python
Product=list(df.Product.unique())

tdf=df.groupby(['Product',"Month"],as_index=False).sum()
g=1
for i in Product:
    fig = plt.figure(figsize=(20,100))
    fig.add_subplot(19,2,g)
    g=g+1
    tdf1=tdf[tdf["Product"]==i]

    sns.barplot(x="Month",y="Sales",data=tdf1).set(title=i)
plt.show()
```

# Sales by Month for each State

```python
City=list(df.City.unique())

tdf=df.groupby(['City',"Month"],as_index=False).sum()
g=1
for i in City:
    fig = plt.figure(figsize=(20,100))
    fig.add_subplot(19,2,g)
    g=g+1
    tdf1=tdf[tdf["City"]==i]
    sns.barplot(x="Month",y="Sales",data=tdf1).set(title=i)
plt.show()
```

# States with most Sales

```python
tdf=df.groupby(['City'],as_index=False).sum()
plt.xticks(rotation=90)
sns.barplot(x="City",y="Sales",data=tdf)
```
